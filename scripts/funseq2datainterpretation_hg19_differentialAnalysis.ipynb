### Data cleaning

```python
# DATA cleaning: 

# first, create SNP_candidates_motifEvents_df by subsetting SNP_candidates_df for columns which motif.analysis has either 'MOTIFBR' or 'MOTIFG' substring
SNP_candidates_motifEvents_df = SNP_candidates_df[SNP_candidates_df['motif.analysis'].str.contains('MOTIFBR|MOTIFG', na=False)] 

print('SNP_candidates_motifEvents_df shape before cleaning',SNP_candidates_motifEvents_df.shape)

# find the rows which contain both 'MOTIFBR' and 'MOTIFG' substrings in motif.analysis column
SNP_candidates_motifBR_and_MOTIFG_df = SNP_candidates_df[SNP_candidates_df['motif.analysis'].str.contains('MOTIFBR', na=False) & SNP_candidates_df['motif.analysis'].str.contains('MOTIFG', na=False)]
# print('SNP_candidates_motifBR_and_MOTIFG_df shape',SNP_candidates_motifBR_and_MOTIFG_df.shape)
# print(SNP_candidates_motifBR_and_MOTIFG_df)

# remove the rows which contain both 'MOTIFBR' and 'MOTIFG' substrings in motif.analysis column from SNP_candidates_motifEvents_df
# this will leave a cleaned dataframe with rows that only contains either 'MOTIFBR' or 'MOTIFG' substring in motif.analysis column
SNP_candidates_motifEvents_df = SNP_candidates_motifEvents_df[~(SNP_candidates_motifEvents_df.index.isin(SNP_candidates_motifBR_and_MOTIFG_df.index))]
print('SNP_candidates_motifEvents_df shape after removing MOTIFBR and MOTIFG both substring rows',SNP_candidates_motifEvents_df.shape)
```

```python
# Add 2 new columns: motif_event and motif_analysis_values

# Initialize empty columns
SNP_candidates_motifEvents_df['motif_event'] = ''
SNP_candidates_motifEvents_df['motif_analysis_values'] = ''

# Iterate through rows and split values or keep original value based on delimiter presence
for index, row in SNP_candidates_motifEvents_df.iterrows():
    split_values = row['motif.analysis'].split(':')
    if len(split_values) == 2:
        SNP_candidates_motifEvents_df.at[index, 'motif_event'] = split_values[0]
        SNP_candidates_motifEvents_df.at[index, 'motif_analysis_values'] = split_values[1]
    else:
        SNP_candidates_motifEvents_df.at[index, 'motif_event'] = row['motif.analysis']
        SNP_candidates_motifEvents_df.at[index, 'motif_analysis_values'] = row['motif.analysis']
```

### Now create 'motif_analysis_value_set'

```python
# creat column 'motif_analysis_value_set' in motifg_rows which contains set of motifs from column 'motif_analysis_values'
motifg_rows['motif_analysis_value_set'] = motifg_rows['motif_analysis_values'].apply(lambda x: set(x.split(','))) 
```

```python
# create column 'motif_analysis_value_set; in motifbr_rows which contains set of motifs from column 'motif_analysis_values'
# note that the value in motif_analysis_values column can be separated by either ',' or '_' or both
# and there's also possibility of the motif being blank
# so in case the set has item '', remove that item from the set
motifbr_rows['motif_analysis_value_set'] = motifbr_rows['motif_analysis_values'].apply(lambda x: set(re.split(r'[,_]', x)))
motifbr_rows['motif_analysis_value_set'] = motifbr_rows['motif_analysis_value_set'].apply(lambda x: x - {''})

```

### Create table of count of motif for each sample. 
each count is one snp
a snp can be associated with multiple motifs

```python
# first, combine the motifg_rows and motifbr_rows dataframes
motifAll_rows = pd.concat([motifg_rows, motifbr_rows], ignore_index=True)
print(motifAll_rows.shape)
```

```python
# create a list out of the set
motifAll_rows['motif_analysis_value_list'] = motifAll_rows['motif_analysis_value_set'].apply(lambda x: list(x))

# create motifAll_sample_df by grouping motifAll_rows by 'sample' and summing the 'motif_analysis_value_list' column
motifAll_sample_df = motifAll_rows.groupby('sample')['motif_analysis_value_list'].sum().reset_index()

print(motifAll_sample_df.shape)
```

```python
from collections import Counter

motif_count_df = pd.DataFrame(motifAll_sample_df['motif_analysis_value_list'].apply(lambda x: Counter(x)).tolist()).fillna(0).astype(int)
```

```python
# join the motif_count_df with the motifAll_sample_df
motifAll_sample_df = pd.concat([motifAll_sample_df, motif_count_df], axis=1)
print(motifAll_sample_df.head(2))
```

### Perform differential analysis

```python
# Add column 'Sample_type' to motifAll_sample_df based on sample names
# if the sample name has substring 'TN' between the first and second underscore, then it's 'TN' sample type
# if the sample name has substring 'R' between the first and second underscore, then it's 'R' sample type
# else it's 'Unknown' sample type
def determine_sample_type(sample_name):
    parts = sample_name.split('_')
    if len(parts) > 2:
        if 'TN' in parts[1]:
            return 'TN'
        elif 'R' in parts[1]:
            return 'R'
    return 'Unknown'
motifAll_sample_df['Sample_type'] = motifAll_sample_df['sample'].apply(determine_sample_type)

# print the number of samples which are TN, R, and Unknown
sample_counts = motifAll_sample_df['Sample_type'].value_counts()
print("Sample counts by type:")
print(sample_counts)
print(f"\nTN: {sample_counts.get('TN', 0)}, R: {sample_counts.get('R', 0)}, Unknown: {sample_counts.get('Unknown', 0)}")

# move 'Sample_type' columns to be the second column
cols = motifAll_sample_df.columns.tolist()
cols.insert(1, cols.pop(cols.index('Sample_type')))
motifAll_sample_df = motifAll_sample_df[cols]
```

```python
# Filter motifs (>=20% non-zero) and run Mann–Whitney U (TN vs R)
from scipy.stats import mannwhitneyu
import numpy as np

# Define the motif columns (as columns which are not base columns)
base_cols = {"sample", "Sample_type", "motif_analysis_value_list"}
motif_columns = [c for c in motifAll_sample_df.columns if c not in base_cols]
# Keep only numeric motif columns
motif_columns = [c for c in motif_columns if np.issubdtype(motifAll_sample_df[c].dtype, np.number)]
# Drop empty motif name if present
motif_columns = [c for c in motif_columns if c != ""]

# Compute non-zero proportion across all samples
# only keep motifs which more than 20% of the samples (both TN and R) are non zero
nonzero_prop = (motifAll_sample_df[motif_columns] > 0).sum() / len(motifAll_sample_df)
keep_motifs = nonzero_prop[nonzero_prop >= 0.20].index.tolist()

print(len(motif_columns))
print(len(keep_motifs))
print(keep_motifs[:10])
```

```python
rows = []
for motif in keep_motifs:
    tn_vals = motifAll_sample_df.loc[motifAll_sample_df["Sample_type"] == "TN", motif]
    r_vals = motifAll_sample_df.loc[motifAll_sample_df["Sample_type"] == "R", motif]

    # Mann–Whitney U test (two-sided)
    try:
        U, p = mannwhitneyu(tn_vals, r_vals, alternative="two-sided", method="auto")
    except TypeError:
        U, p = mannwhitneyu(tn_vals, r_vals, alternative="two-sided")

    rows.append({
        "motif": motif,
        "prop_nonzero": float(nonzero_prop[motif]),
        "n_TN": int((motifAll_sample_df["Sample_type"] == "TN").sum()),
        "n_R": int((motifAll_sample_df["Sample_type"] == "R").sum()),
        "median_TN": float(np.median(tn_vals)),
        "median_R": float(np.median(r_vals)),
        "U_statistic": float(U),
        "p_value": float(p),
    })

mannwhitney_results = pd.DataFrame(rows).sort_values("p_value")
print("Tests run on", len(mannwhitney_results), "motifs")
print(mannwhitney_results.head(10))
```

```python
# Perform Benjamini-Hochberg (FDR) correction
from statsmodels.stats.multitest import multipletests
reject, pvals_corrected, _, _ = multipletests(mannwhitney_results["p_value"], method="fdr_bh")
mannwhitney_results["p_value_FDR"] = pvals_corrected
mannwhitney_results["reject_null"] = reject
print(mannwhitney_results.head(10))
```

```python
# check if any motif has significant difference between TN and R after FDR correction
significant_motifs = mannwhitney_results[mannwhitney_results["reject_null"]]
print("Significant motifs after FDR correction:", significant_motifs.shape[0])
print(significant_motifs)
```
